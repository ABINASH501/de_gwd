```python
from pyspark.sql.functions import upper

# Read data from bronze table
df_invoices = spark.table("bronze_catalog.bronze_stage_schema.invoices")

# Transform region column to uppercase
df_invoices_transformed = df_invoices.withColumn("region", upper("region"))

# Display the transformed data
display(df_invoices_transformed)
```

```python
from pyspark.sql.functions import to_date

# Convert invoice_date column to date type in yyyy-MM-dd format
df_invoices_final = df_invoices_transformed.withColumn(
    "invoice_date", 
    to_date("invoice_date", "yyyy/M/d")
)

# Display the final transformed data
display(df_invoices_final)

# Print schema to verify date type
df_invoices_final.printSchema()
```

```python
df_invoices_final.write.mode("overwrite").saveAsTable("silver_catalog.silver_schema.invoices")
```
