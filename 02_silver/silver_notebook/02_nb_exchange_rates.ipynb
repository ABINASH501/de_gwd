```python
from pyspark.sql.functions import to_date, col

# Read exchange_rates table from bronze_stage_schema
df_exchange_rates = spark.table("bronze_catalog.bronze_stage_schema.exchange_rates")

print("Original schema:")
df_exchange_rates.printSchema()

# Format date column to yyyy-MM-dd
# The date column appears to be in 'yyyy/M/d' format based on the data
df_exchange_rates_formatted = df_exchange_rates.withColumn(
    "date",
    to_date(col("date"), "yyyy/M/d")
)

print("\nFormatted schema:")
df_exchange_rates_formatted.printSchema()

# Display the formatted data
print("\nSample data with formatted date:")
display(df_exchange_rates_formatted)
```

```python
# Save the formatted exchange_rates data to silver catalog
df_exchange_rates_formatted.write.mode("overwrite").saveAsTable("silver_catalog.silver_schema.exchange_rates")

print("✓ Successfully saved exchange_rates table to silver_catalog.silver_schema.exchange_rates")
print(f"Total rows saved: {df_exchange_rates_formatted.count()}")

# Verify the save
df_verify = spark.table("silver_catalog.silver_schema.exchange_rates")
print(f"\nVerification - Row count in silver table: {df_verify.count()}")
print("\nSchema in silver catalog:")
df_verify.printSchema()
print("\nSample from silver catalog:")
df_verify.show(5, truncate=False)
```
