```python
from pyspark.sql.functions import to_date
df_payments = spark.table("bronze_catalog.bronze_stage_schema.payments")
# Convert invoice_date column to date type in yyyy-MM-dd format
df_payments_final = df_payments.withColumn(
    "payment_date", 
    to_date("payment_date", "yyyy/M/d")
)

# Display the final transformed data
display(df_payments_final)

# Print schema to verify date type
df_payments_final.printSchema()
```

```python
df_payments_final.write.mode("overwrite").saveAsTable("silver_catalog.silver_schema.payments")
```
