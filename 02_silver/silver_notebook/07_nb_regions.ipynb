```python
from pyspark.sql.functions import upper

# Read data from bronze table
df_regions = spark.table("bronze_catalog.bronze_stage_schema.regions")

# Transform region_name and country columns to uppercase
df_regions_transformed = df_regions.withColumn("region_name", upper("region_name")) \
                                   .withColumn("country", upper("country"))

# Display the transformed data
display(df_regions_transformed)
```

```python
df_regions_transformed .printSchema()
```

```python
df_regions_transformed .write.mode("overwrite").saveAsTable("silver_catalog.silver_schema.regions")
```
