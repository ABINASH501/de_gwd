```python
%sql
SELECT
    ROUND(SUM(revenue_usd), 2) AS total_revenue_usd
FROM gold_catalog.gold_schema.fact_sales
WHERE invoice_status != 'Cancelled';
```

```python
%sql
SELECT
    SUM(quantity) AS total_quantity_sold
FROM gold_catalog.gold_schema.fact_sales
WHERE invoice_status != 'Cancelled';
```

```python
%sql
WITH invoice_totals AS (
    SELECT
        invoice_id,
        SUM(revenue_usd) AS invoice_value
    FROM gold_catalog.gold_schema.fact_sales
    WHERE invoice_status != 'Cancelled'
    GROUP BY invoice_id
)

SELECT
    ROUND(AVG(invoice_value), 2) AS avg_invoice_value
FROM invoice_totals;
```

```python
%sql
SELECT
    dc.customer_name,
    ROUND(SUM(fs.revenue_usd), 2) AS total_revenue
FROM gold_catalog.gold_schema.fact_sales fs
JOIN gold_catalog.gold_schema.dim_customer dc
    ON fs.customer_key = dc.customer_key
WHERE fs.invoice_status != 'Cancelled'
GROUP BY dc.customer_name
ORDER BY total_revenue DESC
LIMIT 10;
```

```python
%sql
SELECT
    dp.product_name,
    ROUND(SUM(fs.revenue_usd), 2) AS total_revenue
FROM gold_catalog.gold_schema.fact_sales fs
JOIN gold_catalog.gold_schema.dim_product dp
    ON fs.product_key = dp.product_key
WHERE fs.invoice_status != 'Cancelled'
GROUP BY dp.product_name
ORDER BY total_revenue DESC
LIMIT 10;
```

```python
%sql
WITH invoice_status_cte AS (
    SELECT DISTINCT
        invoice_id,
        invoice_status
    FROM gold_catalog.gold_schema.fact_sales
)

SELECT
    ROUND(
        SUM(CASE WHEN invoice_status='Cancelled' THEN 1 ELSE 0 END)
        *100.0 / COUNT(*), 2
    ) AS cancellation_rate_percent
FROM invoice_status_cte;
```

```python
%sql
SELECT
    dc.customer_name,
    ROUND(AVG(fs.discount_amount), 2) AS avg_discount
FROM gold_catalog.gold_schema.fact_sales fs
JOIN gold_catalog.gold_schema.dim_customer dc
    ON fs.customer_key = dc.customer_key
WHERE fs.invoice_status != 'Cancelled'
GROUP BY dc.customer_name
ORDER BY avg_discount DESC;
```

```python
%sql
SELECT
    dr.region_name,
    ROUND(SUM(fs.revenue_usd), 2) AS revenue
FROM gold_catalog.gold_schema.fact_sales fs
JOIN gold_catalog.gold_schema.dim_region dr
    ON fs.region_key = dr.region_key
WHERE fs.invoice_status != 'Cancelled'
GROUP BY dr.region_name
ORDER BY revenue DESC;
```

```python
%sql
SELECT
    dc.customer_name,
    COUNT(DISTINCT fs.invoice_id) AS invoice_count
FROM gold_catalog.gold_schema.fact_sales fs
JOIN gold_catalog.gold_schema.dim_customer dc
    ON fs.customer_key = dc.customer_key
WHERE fs.invoice_status != 'Cancelled'
GROUP BY dc.customer_name
ORDER BY invoice_count DESC;
```

```python
%sql
SELECT
    dd.year,
    dd.month,
    ROUND(SUM(fs.revenue_usd), 2) AS monthly_revenue
FROM gold_catalog.gold_schema.fact_sales fs
JOIN gold_catalog.gold_schema.dim_date dd
    ON fs.date_key = dd.date_key
WHERE fs.invoice_status != 'Cancelled'
GROUP BY dd.year, dd.month
ORDER BY dd.year, dd.month;
```
